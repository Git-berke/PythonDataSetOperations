### İstatistiksel Özet Açıklamaları

- **count**:
  - Her sütundaki toplam veri sayısını gösterir. Bu, o sütunda kaç adet veri noktası olduğunu belirtir.

- **mean**:
  - Her sütundaki değerlerin ortalama değerini gösterir. Bu, merkezi eğilim ölçüsüdür ve verilerin genel eğilimini anlamanıza yardımcı olur.

- **std (standard deviation)**:
  - Standart sapma, her sütundaki değerlerin ne kadar yayıldığını veya dağıldığını gösterir. Yüksek bir standart sapma, değerlerin ortalamadan daha fazla sapma eğiliminde olduğunu gösterir.

- **min**:
  - Her sütundaki en küçük (minimum) değeri gösterir. Bu, veri setindeki alt sınırı belirler.

- **25% (1. çeyrek veya Q1)**:
  - Verilerin %25'inin bu değerin altında olduğunu gösterir. Bu, verilerin alt %25'lik dilimini belirler.

- **50% (medyan veya 2. çeyrek veya Q2)**:
  - Verilerin %50'sinin bu değerin altında olduğunu gösterir. Bu, ortanca değerdir ve verilerin merkezi değerini temsil eder.

- **75% (3. çeyrek veya Q3)**:
  - Verilerin %75'inin bu değerin altında olduğunu gösterir. Bu, verilerin üst %25'lik dilimini belirler.

- **max**:
  - Her sütundaki en büyük (maksimum) değeri gösterir. Bu, veri setindeki üst sınırı belirler.

Tipik anginadan farklı olarak, daha az belirgin veya farklı karakterde olan göğüs ağrısı.
Çeşitli nedenlere bağlı olabilir ve kalp dışı nedenlerle de ilişkili olabilir.

Kalp kaynaklı olmayan göğüs ağrısı.
Sindirim sistemi sorunları, kas-iskelet sistemi rahatsızlıkları veya diğer nedenlerle ortaya çıkabilir.

Göğüs ağrısı belirtisi göstermeyen bireyler.
